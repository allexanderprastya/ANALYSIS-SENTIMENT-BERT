```python
import pandas as pd
import swifter
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize 
```

```python
df = pd.read_csv('/content/review_15k_clean.csv')
normalized_word = pd.read_csv("/content/kamus_alay.csv", sep=";", header=None)
```

```python
df
```

```python
normalized_word
```

```python
def word_tokenize_wrapper(text):
    return word_tokenize(text)

df['content'] = df['content'].swifter.apply(word_tokenize_wrapper)
```

```python
normalized_word_dict = {}

for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]
```

```python
df['content'] = df['content'].swifter.apply(normalized_term) 
```

```python
df['content'] = [' '.join(map(str, l)) for l in df['content']]
```

```python
df.to_csv('review_15k_slang_wordss.csv',  encoding='utf8', index=None, header=True)
```
